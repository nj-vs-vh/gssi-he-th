```python
import astropy.constants as const
import astropy.units as u
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

from plotting import label_axes
```

```python
gamma = 5 / 3
```

```python
fig, axes = plt.subplots(ncols=3, figsize=(10, 3))

M_grid = np.logspace(0, 3, 100)

u_ratio = ((gamma - 1) * M_grid**2 + 2) / ((gamma + 1) * M_grid**2)
P_ratio = 2 * gamma * M_grid**2 / (gamma + 1) - (gamma - 1) / (gamma + 1)
T_ratio = (
    (2 * gamma * M_grid**2 - (gamma - 1))
    * ((gamma - 1) * M_grid**2 + 2)
    / ((gamma + 1) ** 2 * M_grid**2)
)

for ax, label, values in zip(axes, ("u", "P", "T"), (u_ratio, P_ratio, T_ratio)):
    ax.loglog(M_grid, values)
    ax.set_xlabel("$\\mathcal{M}$")
    ax.set_ylabel(f"$ {label}_1 / {label}_2 $")

axes[0].axhline(1/4, color="k", linestyle="--")
plt.tight_layout()
plt.show()
```
