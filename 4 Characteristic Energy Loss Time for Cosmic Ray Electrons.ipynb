# Quadratic energy loss law

Typical energy loss law (e.g. synchrotron and Inverse Compton):

$$
P = \frac{dE}{dt} = - A E^2
$$

Here $A$ is some constant, measured in $\mathrm{E}^{-1} \mathrm{T}^{-1}$.

Solving, we obtain the law for $E(t)$ and introduce characteristic time $\tau \equiv \frac{E(0)}{| P(0) |} = \frac{1}{A E_0} $, so that

$$
E(t) = E_0 \frac{1}{1 + E_0 A t} \\
E(0) \equiv E_0 \\
E(t) = \frac{E_0}{1 + t / \tau}
$$

So, the specific meaning of $\tau$ is that after $t = N\tau$ the initial energy is reduced by factor $\frac 1 {N + 1}$.

```python
A = 1
B = 1
E_0 = 1
tau_1 = 1 / (A * E_0)
tau_2 = 1 / B
t_grid = np.logspace(-0.3, 1.1, 500)

fig, ax = plt.subplots()

E_1 = E_0 / (1 + E_0 * A * t_grid)
ax.plot(t_grid, E_1 / E_0, color="blue", label="$ P \\propto E^{2} $")
ax.plot(t_grid, np.exp(-t_grid / tau_2), color="red", label="$ P \\propto E $")
ax.set_ylabel("$ E / E_0 $")
ax.set_xlabel("$ t $")
for N in range(1, 10):
    ax.axvline(N * tau_1, linestyle="--", color="k")

# ax.set_xlim(left=0)
ax.set_yscale("log")
ax.set_xscale("log")
ax.legend()

plt.show()
```
